```python
import pandas as pd
from pathlib import Path

DATA_DIR = Path("data/raw")

train_tx = pd.read_csv(DATA_DIR / "train_transaction.csv")
train_id = pd.read_csv(DATA_DIR / "train_identity.csv")

print("train_tx:", train_tx.shape)
print("train_id:", train_id.shape)
train_tx.head()
```

```python
df = train_tx.merge(train_id, on="TransactionID", how="left")
print("joined shape:", df.shape)

print("isFraud distribution:")
print(df["isFraud"].value_counts(dropna=False))
print(df["isFraud"].value_counts(normalize=True, dropna=False))

missing_rate = df.isna().mean().sort_values(ascending=False).head(20)
missing_rate
```

```python
print("TransactionDT min/max:", df["TransactionDT"].min(), df["TransactionDT"].max())
print("Unique TransactionIDs:", df["TransactionID"].nunique())
```
